#2. Agrupamiento de los datos

### Modificando el formato del Dataframe

Para realizar las agrupaciones en los datos de emisiones, la base de datos debe estar en un formato adecuado, con los valores de emisiones todos en la misma columna. Para ello debemos modificar el formato de la base de datos para que las columnas referentes a cada ao se conviertan en una nica columna.

```python
emision_gases.loc[:,'Nvel 1 - Setor':'Produto'].columns
```

```python
columnas_info = list(emision_gases.loc[:,'Nvel 1 - Setor':'Produto'].columns)
columnas_info
```

```python
emision_gases.loc[:,1970:2021].columns
```

```python
columnas_emision = list(emision_gases.loc[:,1970:2021].columns)
columnas_emision
```

Para realizar la transformacin del DataFrame, utilizamos el mtodo [`melt()`](https://pandas.pydata.org/docs/reference/api/pandas.melt.html). Este mtodo tiene como objetivo transformar un DataFrame de un formato ancho (*wide*) para un formato largo (*long*).

```python
emision_gases.melt(id_vars = columnas_info, value_vars = columnas_emision, var_name = 'Ao' , value_name = 'Emision')
```

```python
emisiones_por_ao = emision_gases.melt(id_vars = columnas_info, value_vars = columnas_emision, var_name = 'Ao' , value_name = 'Emision')
```

### Anlisis de las emisiones

Vamos a realizar un anlisis, pero esta vez, de la cantidad de la emisin **total** por tipo de gas. Para esto, vamos a realizar un agrupamiento de los datos con la funcin [`groupby()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html).


```python
emisiones_por_ao.groupby('Gs')
```

Podemos revisar cules datos fueron agrupados en cada grupo con el uso del atributo [`groups`](https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.GroupBy.groups.html) y realizar selecciones de grupos especficos con el mtodo [`get_group()`](https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.GroupBy.get_group.html).

```python
emisiones_por_ao.groupby('Gs').groups
```

```python
emisiones_por_ao.groupby('Gs').get_group('CO2 (t)')
```

Los mtodos de agregacin posibles con la utilizacin de `groupby()`son los siguientes:
- `count()` para realizar el conteo
- `sum()` para la suma
- `mean()` para el promedio
- `median()` para la mediana
- `min()` para el valor mnimo
- `max()` para el valor mximo
- `std()` para la desviacin estndar
- `var()` para la varianza

```python
emisiones_por_ao.groupby('Gs').sum()
```

```python
emisiones_por_gas = emisiones_por_ao.groupby('Gs').sum().sort_values('Emision', ascending = False)
emisiones_por_gas
```

```python
emisiones_por_gas.plot(kind = 'barh', figsize = (10,6));
```

```python
emisiones_por_gas.iloc[0:9]
```

```python
print(f'Las emisiones de CO2 corresponden a {float(emisiones_por_gas.iloc[0:9].sum()/emisiones_por_gas.sum())*100:.2f} % de las emisiones totales de gases de efecto invernadero en Brasil de 1970 a 2021.')
```

#3. Agrupamientos multi-index

### Emisin de gases por sector

Ahora necesitamos encontrar dos informaciones:

- Descubrir cuales son los sectores ms contaminantes por cada gs;
- Descubrir cuales son los gases ms contaminantes por cada sector.

```python
gas_por_sector = emisiones_por_ao.groupby(['Gs', 'Nvel 1 - Setor']).sum()
gas_por_sector
```

Para descubrir el sector que emite mayor cantidad de CO2, podemos realizar una seleccin especfica de gas CO2 mediante el mtodo *cross-section* [`xs()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.xs.html), extrayendo de esto el valor mximo y el ndice mximo con los mtodos [`max()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.max.html) y [`idxmax()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.idxmax.html?highlight=idxmax#pandas.DataFrame.idxmax).

```python
gas_por_sector.xs('CO2 (t)', level = 0)
```

```python
gas_por_sector.xs(('CO2 (t)', 'Mudana de Uso da Terra e Floresta'), level = [0,1])
```

```python
gas_por_sector.xs('CO2 (t)', level = 0).max()
```

```python
gas_por_sector.xs('CO2 (t)', level = 0).idxmax()
```

Para evitar un mtodo manual para cada tipo de gas, podemos explorar ms a fondo el mtodo. `groupby()` combinado con [`idxmax()`](https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.DataFrameGroupBy.idxmax.html?highlight=idxmax#pandas.core.groupby.DataFrameGroupBy.idxmax).

```python
gas_por_sector.groupby(level = 0).idxmax()
```

Finalmente, podemos crear una tabla resumida con informacin de los valores mximos de emisin para cada tipo de gas y con la referencia del sector con mayor emisin de cada gas.

```python
gas_por_sector.groupby(level = 0).max()
```

```python
valores_max = gas_por_sector.groupby(level = 0).max().values
```

```python
tabla_totalizada = gas_por_sector.groupby(level = 0).idxmax()
tabla_totalizada.insert(1, 'Cantidad de emisin', valores_max)
tabla_totalizada
```

Podemos cambiar el orden de los niveles del ndice para encontrar la otra informacin, que nos dice cul es el gas ms emitido para cada tipo de sector. Para ello utilizaremos el mtodo [`swaplevel()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.swaplevel.html).

```python
gas_por_sector.swaplevel(0, 1)
```

```python
gas_por_sector.swaplevel(0, 1).groupby(level = 0).idxmax()
```

#4. Uniendo los datos

### Poblacin de los estados

Importaremos una nueva base de datos con informacin de la poblacin de cada estado de Brasil, para luego encontrar la emisin per cpita de cada estado.


Los datos se obtuvieron a travs de [censo del IBGE](https://www.ibge.gov.br/estatisticas/sociais/saude/22827-censo-demografico-2022.html?=&t=resultados)

```python
poblacion_estados = pd.read_excel('POP2022_Municipios.xls', header = 1, skipfooter = 34)
poblacion_estados
```

Como la informacin se refiere a municipios, realizaremos una agregacin basada en la columna estado.

```python
poblacion_estados.groupby('UF').sum(numeric_only = True)
```

Notamos que la informacin de la poblacin no estaba agregada, por lo que ser necesario realizar algn tratamiento en esta columna para devolver los datos correctamente.

```python
#poblacion_estados['POPULAO'].astype(int)
```

```python
poblacion_estados[poblacion_estados['POPULAO'].str.contains('\(', na = False)]
```

```python
poblacion_estados = poblacion_estados.assign(poblacion_sin_parentesis = poblacion_estados['POPULAO'].replace('\(\d{1,2}\)', '', regex = True),
                                             poblacion = lambda x: x.loc[:,'poblacion_sin_parentesis'].replace('\.', '', regex = True))
```

```python
poblacion_estados[poblacion_estados['POPULAO'].str.contains('\(', na = False)]
```

```python
poblacion_estados.loc[:,'poblacion'] = poblacion_estados['poblacion'].astype(int)
```

```python
poblacion_estados = poblacion_estados.groupby('UF').sum(numeric_only = True)['poblacion'].reset_index()
poblacion_estados
```

### Uniendo los datos

Con la informacin procesada podemos unir los datos de emisiones con los datos de la poblacin, utilizando el mtodo [`merge()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html).

```python
emision_estados = emisiones_por_ao[emisiones_por_ao['Ao'] == 2021].groupby('Estado').sum(numeric_only = True).reset_index()
emision_estados
```

```python
datos_agrupados = pd.merge(emision_estados, poblacion_estados, left_on = 'Estado', right_on = 'UF')
datos_agrupados
```

Comprobemos si hay algn indicio de si las emisiones aumentan con el tamao de la poblacin.

```python
datos_agrupados.plot(x = 'poblacion', y= 'Emision', kind = 'scatter', figsize=(8,6));
```

Si queremos conocer la informacin de cada estado en los puntos del grfico, podemos utilizar otra biblioteca para obtener un grfico ms elaborado.

```python
import plotly.express as px
```

```python
px.scatter(data_frame = datos_agrupados, x = 'poblacion', y = 'Emision', text = 'Estado', opacity = 0)
```

Para obtener las emisiones per cpita crearemos una nueva columna dividiendo los valores de emisin por la poblacin de cada estado.

```python
datos_agrupados = datos_agrupados.assign(emision_per_capita = datos_agrupados['Emision']/datos_agrupados['poblacion']).sort_values('emision_per_capita', ascending = False)
datos_agrupados
```

```python
px.bar(data_frame = datos_agrupados, x = 'Estado', y = 'emision_per_capita')
```

```python
px.scatter(data_frame = datos_agrupados, x = 'poblacion', y = 'Emision', text = 'Estado', size = 'emision_per_capita')
```
